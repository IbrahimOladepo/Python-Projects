```python
plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('Temparature in degree celcius')
plt.ylabel('Rainfall in meters cube')
plt.title('A graph of rainfall against temparature')
```

## Crating multiplot on same canvas

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'ko')

plt.subplot(1, 2, 2)
plt.plot(X,Y,'g+')
```

```python
# create figure
fig = plt.figure()

# add set of axes to figure
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# plot on that set of axes
ax.set_xlabel('Temparature')
ax.set_ylabel('Rainfall')
ax.set_title('Rainfall against Temparature')
ax.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
# create figure
fig = plt.figure()

# add set of axes to figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.4, 0.2, 0.4, 0.4])


# plot on that set of axes
ax1.set_xlabel('Temparature')
ax1.set_ylabel('Rainfall')
ax1.set_title('Rainfall against Temparature')
ax1.plot(x, y, 'r+')

ax2.set_xlabel('Water')
ax2.set_ylabel('Air')
ax2.set_title('Air against Water')
ax2.plot(x, y, 'b')
```
